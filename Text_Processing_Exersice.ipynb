```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import re
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
```

```python
df = pd.read_csv('/content/tripadvisor_hotel_reviews.csv')
conditions = [
    (df['Rating'] <= 2),
    (df['Rating'] ==3),
    (df['Rating'] > 3) 
    ]

# create a list of the values we want to assign for each condition
values = [0,1,2]

# create a new column and use np.select to assign values to it using our lists as arguments
df['Rating_numeric'] = np.select(conditions, values)
df['Review'] = df['Review'].str.lower()
#df['Review']=re.sub('[^a-z]+', ' ', df['Review'])
df = df.replace(to_replace ='[^a-z]', value = ' ', regex = True)
df['Review']=df['Review'].str.replace("  "," ")

stop = stopwords.words('english')
df['Review_without_stopwords'] = df['Review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df['Review_tokenized'] = df.apply(lambda row: nltk.word_tokenize(row["Review_without_stopwords"]), axis=1)
ps = PorterStemmer()
df['Review_stemmed'] = df['Review_tokenized'].apply(lambda x: [ps.stem(y) for y in x])
df['Review']= df['Review_stemmed'].str.join(" ")
df
```
